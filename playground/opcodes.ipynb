```python
import json
import os

# Construct the path to the JSON file
json_file_path = os.path.join(os.getcwd(), '../assets/opcode-table.json')

# Read the JSON file
with open(json_file_path, 'r') as file:
    data = json.load(file)


bytes = [
    {
        "name": " ".join(x["bytes"]),
        "opcode": x,
    }
    for x in data
]

bytes.sort(key=lambda x: x["name"])

with open(os.path.join(os.getcwd(), './sorted_opcode_table.json'), 'w') as outfile:
    json.dump(bytes, outfile, indent=4)

# display(bytes)

names = [x["name"] for x in bytes]
display(names)

with open(os.path.join(os.getcwd(), './sorted_opcodes.txt'), 'w') as outfile:
    for n in names:
        outfile.write(n + '\n')
```
